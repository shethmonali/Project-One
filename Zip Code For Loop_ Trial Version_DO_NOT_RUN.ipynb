```python
# create a params dict that will be updated with new city each iteration
params = {"key": yelp_key}
# Loop through the cities_pd and run a lat/long search for each city
for index, row in zip_code_pd.iterrows():
    url = "https://api.yelp.com/v3/businesses/search?term=food&location=%s" % ("row")
    # update address key value
    params['location'] = f"{row['Zip Code']}"
    #params['businesses'] = f"{row['Search Term']}"
    # make request, print url
    zip_code_yelp = requests.get(url, params=params)
    print(zip_code_yelp.url)
    # convert to json
    zip_code_yelp = zip_code_yelp.json()
    try:
        zip_code_pd.at[index, "Zip Code"]=zip_code_yelp["location"]["zip_code"]
        #zip_code_pd.at[index, "Lng"]=cities_lat_lng["results"][0]["geometry"]["location"]["lng"]
    except (KeyError, IndexError) as e:
        print(f"error at {row['Zip Code']}, dropping zip code: {e}")
        #remove row for city that doesn't have coordinates
        zip_code_pd.drop(index,axis=0,inplace=True)
        
# Visualize to confirm lat lng appear
zip_code_pd.head()
```
